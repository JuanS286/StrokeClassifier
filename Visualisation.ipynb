```python
import matplotlib.pyplot as plt

# Metrics for SVM and RF
svm_metrics = {
    "Accuracy": 0.9598650927487352,
    "Precision": 0.9819851642529142,
    "F1-score": 0.9589513625388065,
    "Recall": 0.9369733737782272
}

rf_metrics = {
    "Accuracy": 0.9812739604836331,
    "Precision": 0.9638999431495168,
    "F1-score": 0.9816181791865682,
    "Recall": 1.0
}

# Data preparation
metrics = list(svm_metrics.keys())
svm_values = list(svm_metrics.values())
rf_values = list(rf_metrics.values())

# Creating the plot with a larger figure size
fig, ax = plt.subplots(figsize=(10, 8))
bar_width = 0.35
index = range(len(metrics))

bar1 = ax.bar(index, svm_values, bar_width, label='SVM', color='lightgreen')
bar2 = ax.bar([p + bar_width for p in index], rf_values, bar_width, label='RF', color='orange')

ax.set_xlabel('Metrics')
ax.set_ylabel('Values')
ax.set_title('Comparison of SVM and RF Training Metrics')
ax.set_xticks([p + bar_width / 2 for p in index])
ax.set_xticklabels(metrics)

# Place the legend outside the plot
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Adding values above the bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('%.2f' % height,
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords='offset points',
                    ha='center', va='bottom')

add_labels(bar1)
add_labels(bar2)

plt.tight_layout()  # Adjust layout to make room for the legend
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion matrices data
rf_cm = np.array([[797, 48], [23, 21]])
svm_cm = np.array([[622, 20], [43, 586]])

# Swapping the order of the confusion matrices for display
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

sns.heatmap(svm_cm, annot=True, fmt="d", cmap="Greens", ax=ax[0])
ax[0].set_title('SVM Test Metrics')
ax[0].set_xlabel('Predicted Labels')
ax[0].set_ylabel('True Labels')
ax[0].set_xticklabels(['Negative', 'Positive'])
ax[0].set_yticklabels(['Negative', 'Positive'])

sns.heatmap(rf_cm, annot=True, fmt="d", cmap="Oranges", ax=ax[1])
ax[1].set_title('Random Forest Test Metrics')
ax[1].set_xlabel('Predicted Labels')
ax[1].set_ylabel('True Labels')
ax[1].set_xticklabels(['Negative', 'Positive'])
ax[1].set_yticklabels(['Negative', 'Positive'])

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Updated metrics for SVM and RF on test data
svm_test_metrics = {
    "Test Accuracy": 0.955153422501967,
    "Test Precision": 0.9724473257698542
}

rf_test_metrics = {
    "Test Accuracy": 0.9201349831271091,
    "Test Precision": 0.30434782608695654
}

# Data preparation
test_metrics = list(svm_test_metrics.keys())
svm_test_values = list(svm_test_metrics.values())
rf_test_values = list(rf_test_metrics.values())

# Creating the bar chart
fig, ax = plt.subplots(figsize=(8, 6))
bar_width = 0.35
index = range(len(test_metrics))

bar1 = ax.bar(index, svm_test_values, bar_width, color='lightgreen', label='SVM')
bar2 = ax.bar([p + bar_width for p in index], rf_test_values, bar_width, color='orange', label='RF')

ax.set_xlabel('Metrics')
ax.set_ylabel('Values')
ax.set_title('Comparison of SVM and RF Test Metrics')
ax.set_xticks([p + bar_width / 2 for p in index])
ax.set_xticklabels(test_metrics)
ax.legend()

# Adding numbers on top of the bars
def add_numbers_on_top(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{:.4f}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

add_numbers_on_top(bar1)
add_numbers_on_top(bar2)

plt.show()
```
